**Linear and Non-Linear Regression:**
 Regression with non-linearity assumption gives superior result because Linear regression is a subset of non-linear regression. But even non linear regression can be enhanced by introducing multi-variant bin variable. This multi-variant bin variables can be generated by plotting a simple decision tree.
